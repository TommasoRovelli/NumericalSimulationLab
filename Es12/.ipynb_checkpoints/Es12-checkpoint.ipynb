## Exercise 12

In this Jupyter Notebook I created a DNN and a CNN to recognize the digit written on a image 28x28. To train the networks I used the MINST dataset that comprises 70000 handwritten digits.

```python
import tensorflow as tf
from tensorflow import keras
import os
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
import os

from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import SGD, Adam, RMSprop, Adagrad, Adadelta, Adam, Adamax, Nadam
from keras.layers import Flatten, Conv2D, MaxPooling2D

seed=0
np.random.seed(seed) # fix random seed
tf.random.set_seed(seed)
```

```python
# input image dimensions
img_rows, img_cols = 28, 28 # number of pixels 
# output
num_classes = 10 # 10 digits

# the data, split between train and test sets
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()
```

```python
# reshape data, it could depend on Keras backend
X_train = X_train.reshape(X_train.shape[0], img_rows*img_cols)
X_test = X_test.reshape(X_test.shape[0], img_rows*img_cols)

# cast floats to single precision
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

# rescale data in interval [0,1]
X_train /= 255
X_test /= 255

# convert class vectors to binary class matrices, e.g. for use with categorical_crossentropy
Y_train = keras.utils.to_categorical(Y_train, num_classes)
Y_test = keras.utils.to_categorical(Y_test, num_classes)
```

## <span style="color:green"> Deep Neural Network </span>

```python
def create_DNN():
    # instantiate model
    model = Sequential()
    # add a dense all-to-all relu layer
    model.add(Dense(400,input_shape=(img_rows*img_cols,), activation='relu'))
    # add a dense all-to-all relu layer
    model.add(Dense(100, activation='relu'))
    # apply dropout with rate 0.5
    model.add(Dropout(0.5))
    # soft-max layer
    model.add(Dense(num_classes, activation='softmax'))
    
    return model
```

```python
def compile_model(optimizer):
    # create the model
    model=create_DNN()
    # compile the model
    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer=optimizer,
                  metrics=['acc'])
    return model
```

```python
# training parameters
batch_size = 32
epochs = 15

# create the deep neural net
model_DNN_sgd = compile_model('sgd')
model_DNN_adam = compile_model('adam')
model_DNN_adadelta = compile_model('adadelta')

model_DNN_sgd.summary()
```

I have created three deep neural network as descripted in this summary. Each DNN have a different optimizer: I used sgd, adam and adadelta.

```python
history_sgd = model_DNN_sgd.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=0,
          validation_data=(X_test, Y_test))
```

```python
history_adam = model_DNN_adam.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=0,
          validation_data=(X_test, Y_test))
```

```python
history_adadelta = model_DNN_adadelta.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=0,
          validation_data=(X_test, Y_test))
```

```python
fig, ((sgd_loss_plot, adam_loss_plot, adadelta_loss_plot),(sgd_accuracy_plot, adam_accuracy_plot, adadelta_accuracy_plot)) = plt.subplots(2, 3, figsize=(20,10))

min_sgd = min(history_sgd.history['val_loss'])
min_adam = min(history_adam.history['val_loss'])
min_adadelta = min(history_adadelta.history['val_loss'])

max_sgd = max(history_sgd.history['val_acc'])
max_adam = max(history_adam.history['val_acc'])
max_adadelta = max(history_adadelta.history['val_acc'])

# summarize history for loss
sgd_loss_plot.plot(history_sgd.history['loss'])
sgd_loss_plot.plot(history_sgd.history['val_loss'])
sgd_loss_plot.set_ylabel('model loss')
sgd_loss_plot.set_xlabel('epoch')
sgd_loss_plot.set_title("Optimizer: sgd, min loss=%f" %min_sgd)
sgd_loss_plot.legend(['train', 'test'], loc='best')
sgd_loss_plot.grid(True)

# summarize history for loss
adam_loss_plot.plot(history_adam.history['loss'])
adam_loss_plot.plot(history_adam.history['val_loss'])
adam_loss_plot.set_ylabel('model loss')
adam_loss_plot.set_xlabel('epoch')
adam_loss_plot.set_title("Optimizer: adam, min loss=%f" %min_adam)
adam_loss_plot.legend(['train', 'test'], loc='best')
adam_loss_plot.grid(True)

# summarize history for loss
adadelta_loss_plot.plot(history_adadelta.history['loss'])
adadelta_loss_plot.plot(history_adadelta.history['val_loss'])
adadelta_loss_plot.set_ylabel('model loss')
adadelta_loss_plot.set_xlabel('epoch')
adadelta_loss_plot.set_title("Optimizer: adadelta, min loss=%f" %min_adadelta)
adadelta_loss_plot.legend(['train', 'test'], loc='best')
adadelta_loss_plot.grid(True)

# summarize history for accuracy
sgd_accuracy_plot.plot(history_sgd.history['acc'])
sgd_accuracy_plot.plot(history_sgd.history['val_acc'])
sgd_accuracy_plot.set_ylabel('model acccuracy')
sgd_accuracy_plot.set_xlabel('epoch')
sgd_accuracy_plot.set_title("Optimizer: sgd, max accuracy=%f" %max_sgd)
sgd_accuracy_plot.legend(['train', 'test'], loc='best')
sgd_accuracy_plot.grid(True)

# summarize history for accuracy
adam_accuracy_plot.plot(history_adam.history['acc'])
adam_accuracy_plot.plot(history_adam.history['val_acc'])
adam_accuracy_plot.set_ylabel('model accuracy')
adam_accuracy_plot.set_xlabel('epoch')
adam_accuracy_plot.set_title("Optimizer: adam, max accuracy=%f" %max_adam)
adam_accuracy_plot.legend(['train', 'test'], loc='best')
adam_accuracy_plot.grid(True)

# summarize history for accuracy
adadelta_accuracy_plot.plot(history_adadelta.history['acc'])
adadelta_accuracy_plot.plot(history_adadelta.history['val_acc'])
adadelta_accuracy_plot.set_ylabel('model loss')
adadelta_accuracy_plot.set_xlabel('epoch')
adadelta_accuracy_plot.set_title("Optimizer: adadelta, max accuracy=%f" %max_adadelta)
adadelta_accuracy_plot.legend(['train', 'test'], loc='best')
adadelta_accuracy_plot.grid(True)

plt.show()
```

As shown in those graphs adadelta is the worst (probably I need to adjust some parameters to make it works in the right way). <br>
Adam shows a minimum of the test loss even if the train loss continue to decrease. This happen because the DNN overfit the train data.

The best optimizer I used is clearly Adam because reach very low loss with few epoches. The only precaution to bear in mind is to stop the training at the right time and not let the network overfit the training data.
                                                 
In all the graphs the training loss is higher than the test one. This behaviour follows from the presence of a dropout layer.

```python
X_test = X_test.reshape(X_test.shape[0], img_rows*img_cols)

predictions_sgd = model_DNN_sgd.predict(X_test)
predictions_adam = model_DNN_sgd.predict(X_test)
predictions_adadelta= model_DNN_adadelta.predict(X_test)

X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols,1)

plt.figure(figsize=(15, 15)) 

for i in range(10):    
    ax1 = plt.subplot(2, 10, i + 1)    
    plt.imshow(X_test[i, :, :, 0], cmap='gray')    
    plt.title("Digit: {}\nSgd: {}\nAdam: {}\nAdadelta: {}".format(np.argmax(Y_test[i]), np.argmax(predictions_sgd[i]), np.argmax(predictions_adam[i]), np.argmax(predictions_adadelta[i])))    
    plt.axis('off') 
    
plt.show()
```

This is the explicit representation of the DNN work. Adam and sgd have guessed all the digits instead adadelta guessed all but one. This behaviour is in accord with the accuracy of the networks (sgd:97.3\%, adam: 98.23\%, adadelta: 85.94\%).

## <span style="color:green"> Convolutional Neural Network </span>

A convolutional neural network contains some convolutional layers and some pooling layers. The purpose of those layers is to compress and isolate some tipical traits of the images to improve the recognitions of the digits. <br>

For this CNN I used sgd as optimizer.

```python
if keras.backend.image_data_format() == 'channels_first':
    X_train = X_train.reshape(X_train.shape[0], 1, img_rows, img_cols)
    X_test = X_test.reshape(X_test.shape[0], 1, img_rows, img_cols)
    input_shape = (1, img_rows, img_cols)
else:
    X_train = X_train.reshape(X_train.shape[0], img_rows, img_cols, 1)
    X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols, 1)
    input_shape = (img_rows, img_cols, 1)
```

```python
def create_CNN():
    # instantiate model
    model = Sequential()
    # add first convolutional layer with 10 filters (dimensionality of output space)
    
    model.add(Conv2D(10, kernel_size=(5, 5), input_shape=input_shape, activation='relu'))
    
    model.add(MaxPooling2D(pool_size=(2,2), padding="valid"))
    
    model.add(Conv2D(16, kernel_size=(5, 5), activation='relu'))
    
    model.add(MaxPooling2D(pool_size=(2,2), strides=1, padding="valid"))
    
    model.add(Flatten())
    
    model.add(Dense(100, activation='relu'))
    model.add(Dense(num_classes, activation='softmax'))
    
    # compile the model
    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer='SGD',
                  metrics=['acc'])
    return model
```

```python
def compile_model(optimizer):
    # create the model
    model=create_CNN()
    # compile the model
    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer=optimizer,
                  metrics=['acc'])
    return model
```

```python
batch_size = 32
epochs = 10

model = compile_model('sgd')
```

```python
model.summary()
```

I created a convolutiona neural network as descripted in this summary.

```python
history = model.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(X_test, Y_test))
```

```python
min_loss = min(history.history['val_loss'])
max_acc = max(history.history['val_acc'])

fig, (plot_loss, plot_acc) = plt.subplots(1,2,figsize=(20,5))

plot_loss.plot(history.history['loss'])
plot_loss.plot(history.history['val_loss'])
plot_loss.set_ylabel('model loss')
plot_loss.set_xlabel('epoch')
plot_loss.set_title("Loss over epoches, min loss=%f" %min_loss)
plot_loss.legend(['train', 'test'], loc='best')
plot_loss.grid(True)

plot_acc.plot(history.history['acc'])
plot_acc.plot(history.history['val_acc'])
plot_acc.set_ylabel('model accuracy')
plot_acc.set_xlabel('epoch')
plot_acc.set_title("Accuracy over epoches, max accuracy=%f" %max_acc)
plot_acc.legend(['train', 'test'], loc='best')
plot_acc.grid(True)

plt.show()
```

The loss decrease more quickly than the DNN one and only after 10 epoches the CNN reaches an accuracy of 98.75\%.

```python
if keras.backend.image_data_format() == 'channels_first':
    X_test = X_test.reshape(X_test.shape[0], 1, img_rows, img_cols)
else:
    X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols, 1)
    
predictions = model.predict(X_test)

X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols,1)

plt.figure(figsize=(25, 15)) 
for i in range(15):    
    ax = plt.subplot(2, 15, i + 1)    
    plt.imshow(X_test[i, :, :, 0], cmap='gray')    
    plt.title("Digit: {}\nPredicted:  {}".format(np.argmax(Y_test[i]), np.argmax(predictions[i])))    
    plt.axis('off') 
plt.show()
```

For the digits of the MINST set the CNN works admirably.

## <span style="color:green"> My Handwriting </span>

```python
digits_filenames = np.array([
    "./Digits/" + str(i)+'.png'
    for i in range(10)
])

digit_in = []
pix = []
data = []

for i in range(10):
    im = Image.open(digits_filenames[i]).convert('L')
    digit_in.append(im)

ydim, xdim = digit_in[0].size

for i in range(10):
    data.append(np.zeros((xdim,ydim)))
    pix.append(digit_in[i].load())

for i in range(10):
    for j in range(xdim):
        for k in range(ydim):
            data[i][j,k] = pix[i][k,j]/255
```

```python
predict_CNN = []
predict_DNN = []

for i in range(10):
    data[i] = data[i].reshape(1,xdim,ydim,1)
    
for i in range(10):
    a = model.predict(data[i])
    predict_CNN.append(a)
    
for i in range(10):
    data[i] = data[i].reshape(data[i].shape[0],xdim*ydim)
    
for i in range(10):
    a = model_DNN_sgd.predict(data[i])
    predict_DNN.append(a)
    
for i in range(10):
    data[i] = data[i].reshape(xdim,ydim)


plt.figure(figsize=(15, 15)) 
for i in range(10):    
    ax = plt.subplot(2, 10, i + 1)    
    plt.imshow(data[i], cmap='gray')
    plt.title("Digit: {}\nCNN: {}\nDNN: {}".format(i, np.argmax(predict_CNN[i]), np.argmax(predict_DNN[i])))
    plt.axis('off')
plt.show()
```

I compared the prevision of the CNN and the DNN (the one optimized with sgd). <br>
Both the neural networks fail several times due to they was trained on the MINST set and my handwriting could not follow the standard of the set.<br>

This shows how a neural network can have troubles to work on a data set slightly different from the training one.
