# Exercise 3
## <span style="color:green"> European Option</span>

The aim here was to establish what kind of price should an european option have both in the call and in the put option. We also did a comparison with the epected values, because luckily - due to the extreme semplicity of the european option in comparison to other option - there is an analytic solution to the problem.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.optimize import curve_fit
```

```python
path_dir_call = "3.1/output_dir_call_3.1.out"
path_step_call = "3.1/output_step_call_3.1.out"
path_dir_put = "3.1/output_dir_put_3.1.out"
path_step_put = "3.1/output_step_put_3.1.out"

M_dc, N_dc = np.loadtxt(path_dir_call, usecols=(0,1), delimiter=' ', unpack='true', max_rows=1)
M_sc, N_sc = np.loadtxt(path_step_call, usecols=(0,1), delimiter=' ', unpack='true', max_rows=1)
M_dp, N_dp = np.loadtxt(path_dir_put, usecols=(0,1), delimiter=' ', unpack='true', max_rows=1)
M_sp, N_sp = np.loadtxt(path_step_put, usecols=(0,1), delimiter=' ', unpack='true', max_rows=1)

ave_dc, err_dc = np.loadtxt(path_dir_call, usecols=(0,1), delimiter=' ', unpack='true', skiprows=1)
ave_sc, err_sc = np.loadtxt(path_step_call, usecols=(0,1), delimiter=' ', unpack='true', skiprows=1)
ave_dp, err_dp = np.loadtxt(path_dir_put, usecols=(0,1), delimiter=' ', unpack='true', skiprows=1)
ave_sp, err_sp = np.loadtxt(path_step_put, usecols=(0,1), delimiter=' ', unpack='true', skiprows=1)

L_dc = int(M_dc/N_dc)
L_sc = int(M_sc/N_sc)
L_dp = int(M_dp/N_dp)
L_sp = int(M_sp/N_sp)

x_dc = np.arange(N_dc)
x_sc = np.arange(N_sc)
x_dp = np.arange(N_dp)
x_sp = np.arange(N_sp)

x_dc*=L_dc
x_sc*=L_sc
x_dp*=L_dp
x_sp*=L_sp

fig, (c_plot,p_plot) = plt.subplots(1,2 , figsize=(26,5))
c_plot.errorbar(x_dc, ave_dc, yerr=err_dc, label="Discrete Sampling", alpha = 0.6)
c_plot.errorbar(x_sc, ave_sc, yerr=err_sc, label="Direct Sampling", alpha = 0.6)
c_plot.axhline(y=14.975790778311286, color="r", label="Theoretical", alpha = 0.6)
c_plot.grid(True)
c_plot.set_title("Call")
c_plot.set_xlabel("#throws")
c_plot.set_ylabel("<C[S(T),T]>")
c_plot.legend()

p_plot.errorbar(x_dp, ave_dp, yerr=err_dp, label="Discrete Sampling", alpha = 0.6)
p_plot.errorbar(x_sp, ave_sp, yerr=err_sp, label="Direct Sampling", alpha=0.6)
p_plot.axhline(y=5.4595325819072364, color="r", label="Theoretical", alpha = 0.6)
p_plot.grid(True)
p_plot.set_title("Put")
p_plot.set_xlabel("#throws")
p_plot.set_ylabel("<P[S(T),T]>")
p_plot.legend()

plt.show()
```

The results of the simulation is in accord with the theoretical results both in the case of:
1. sampling directly the final asset price $S(T)$</span> for a $GBM(r,\sigma^2)$ (blue line)  
2.sampling the discretized $GBM(r,\sigma^2)$ path of the asset price dividing $[0,T]$ in $10^2$ time intervals: $S(0)$, $S(t_1)$, $S(t_2)$, ... up to $S(t_{100}=T=1)$ (orange line)

Obviously the error are smaller using the discrete sampling because it gives a much better prediction of what the value of the good considered is goint to be at the given time T.
