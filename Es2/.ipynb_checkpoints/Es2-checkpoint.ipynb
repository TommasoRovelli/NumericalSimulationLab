### <span style="color:green"> Importance Sampling</span>

In this exercise we compute the integral of $ \dfrac{\pi}{2} \cos( \dfrac{\pi x}{2} ) $ using the Importance Sampling method.

It consist in rewrite the integral in the form $ \dfrac{f(x)}{ \rho(x)} \rho(x)$ when $\rho(x)$ is a probability distribution and then mediate the obtained values of $\dfrac{f(x)}{ \rho(x)}$ with $x$ obtained via $\rho(x)$ (for the mean method we use $\rho(x)=1$). <br>
More similar are $\rho(x)$ and $ f(x) $ better is the results and smaller the error.


```python
path_UD = "2.1/output_UD_2.1.out"
path_IS = "2.1/output_IS_2.1.out"
path_WS = "2.1/output_WS_2.1.out"

N_UD, L_UD = np.loadtxt(path_UD, usecols=(0,1), delimiter=' ', unpack='true', max_rows=1)
N_IS, L_IS = np.loadtxt(path_IS, usecols=(0,1), delimiter=' ', unpack='true', max_rows=1)
N_WS, L_WS = np.loadtxt(path_WS, usecols=(0,1), delimiter=' ', unpack='true', max_rows=1)

ave_UD, err_UD = np.loadtxt(path_UD, usecols=(0,1), delimiter=' ', unpack='true', skiprows=1)
ave_IS, err_IS = np.loadtxt(path_IS, usecols=(0,1), delimiter=' ', unpack='true', skiprows=1)
ave_WS, err_WS = np.loadtxt(path_WS, usecols=(0,1), delimiter=' ', unpack='true', skiprows=1)

x_UD = np.arange(N_UD)
x_UD*=L_UD

x_IS = np.arange(N_IS)
x_IS*=L_IS

x_WS = np.arange(N_WS)
x_WS*=L_WS

plt.title("Extimation of $ \int_0^1 \dfrac{\pi}{2} \cos( \dfrac{\pi x}{2}) dx $ (true value=1)")

UD_plot = plt.errorbar(x_UD,ave_UD,yerr=err_UD)
UD_plot.set_label("Uniform Distribution")

IS_plot = plt.errorbar(x_IS,ave_IS,yerr=err_IS)
IS_plot.set_label("Importance Sampling $\\rho(x)=2(1-x)$")

WS_plot = plt.errorbar(x_WS,ave_WS,yerr=err_WS,color="tab:olive")
WS_plot.set_label("Wrong Sampling $\\rho(x)= \dfrac{x}{2} $")

plt.xlabel('#extimations')
plt.ylabel('<I>')

fig = matplotlib.pyplot.gcf()
fig.set_size_inches(10, 6)

plt.grid(True)
plt.legend(loc=2, prop={'size': 12})

plt.show()
```

It is trivial to observe that the Important Sampling method makes the integral converge faster and the error is smaller than the Mean method. On the other hand, if we use a distribution very different from the intagrand the errors of the extimation are bigger.

### <span style="color:green"> Random Walk </span>

Here we want to compute the variance of a 2D Random Walk both in a cubic lattice and in the continuum. We also want to check its diffusive behaviour.

```python
def sqrt(x,k):
    return k*np.sqrt(x)
```

```python
path_lattice = "2.2/output_lat_2.2.out"
path_continuum = "2.2/output_con_2.2.out"

n_step_lat = np.loadtxt(path_lattice, usecols=(0), delimiter=' ', unpack='true', max_rows=1)
n_step_cont = np.loadtxt(path_continuum, usecols=(0), delimiter=' ', unpack='true', max_rows=1)

ave_lat, err_lat = np.loadtxt(path_lattice, usecols=(0,1), delimiter=' ', unpack='true', skiprows=1)
ave_cont, err_cont = np.loadtxt(path_continuum, usecols=(0,1), delimiter=' ', unpack='true', skiprows=1)

x_lat = np.arange(n_step_lat) 
x_cont = np.arange(n_step_cont) 

popt_lat, pcov_lat = curve_fit(sqrt, xdata=x_lat, ydata=ave_lat)
popt_cont, pcov_cont = curve_fit(sqrt, xdata=x_cont, ydata=ave_cont)

x_err_lat = np.arange(n_step_lat-1) 
x_err_cont = np.arange(n_step_cont-1) 

perc_err_lat = 100*np.divide(err_lat[1:], ave_lat[1:])
perc_err_cont = 100*np.divide(err_cont[1:], ave_cont[1:])

fig1, (lat_plot, cont_plot) = plt.subplots(1, 2 , figsize=(20,6))

lat_plot.errorbar(x_lat,ave_lat,yerr=err_lat, label="Computed")
lat_plot.plot(x_lat, sqrt(x_lat, *popt_lat), color='r',lw=2, label="Fit: f(x)= k*$\sqrt{x}$ ")
lat_plot.set_title("Random Walk on a Cubic Lattice")
lat_plot.legend()

cont_plot.errorbar(x_cont,ave_cont,yerr=err_cont, label="Computed")
cont_plot.plot(x_cont, sqrt(x_cont, *popt_cont), color='r',lw=2, label="Fit: f(x)= k*$\sqrt{x}$ ")
cont_plot.set_title("Random Walk in the Continuum")
cont_plot.legend()

lat_plot.set_xlabel('#step')
lat_plot.set_ylabel('$\sqrt{ < | r_N |^2 >_{RW}}$')

cont_plot.set_xlabel('#step')
cont_plot.set_ylabel('$\sqrt{ < | r_N |^2  >_{RW}}$')

lat_plot.grid(True)
cont_plot.grid(True)


print("  Optimized parameter (cubic lattice): ", "k=",popt_lat[0],"\t\t\t","Optimized parameter (continuum): ", "k=",popt_cont[0])
print("  Parameter uncertainty (cubic lattice): ", np.sqrt(np.diagonal(pcov_lat)), "\t\t\t\t","Parameter uncertainty (continuum): ", np.sqrt(np.diagonal(pcov_cont)) ) 

plt.show()
```

The graph of the $\sqrt{<|r_N|^2>_{RW}}$ follows nearly perfectly the function $\sqrt{N}$ in both cases. <br>
This behaviour shows that the points after N steps are distributed with a density function that follows the diffusive equation 
$\dfrac{\partial \rho}{\partial t} = D\nabla^2\rho$ (in our case $D=\dfrac{a^2}{2\tau}$ and $a=1$ is the step lenght). <br>
Solving the equation with initial conditions $\vec{x}(0)=\vec{0}$ we obtain a origin centered gaussian with standard deviation $\sigma = \sqrt{2Dt} = \sqrt{2\dfrac{a^2}{2\tau}N\tau}=\sqrt{N}$ that is the showed behaviour.
