### <span style="color:green"> Random Number Generator Test </span>

In this exercise we test the random number generator given, computing the mean, the standard deviation and the $\chi^2$ distribution for values extracted in [0,1).

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.optimize import curve_fit
from scipy.stats import chi2
import seaborn as sns
```

```python
path = "1.1/output_ave_1.1.out"

M, N = np.loadtxt(path, usecols=(0,1), delimiter=' ', unpack='true', max_rows=1)

ave, err = np.loadtxt(path, usecols=(0,1), delimiter=' ', unpack='true', skiprows=1)
  
L=int(M/N)

x = np.arange(N)
x*=L 

plt.errorbar(x,ave,yerr=err)
plt.xlabel('#throws')
plt.ylabel('<r>')
plt.title("Mean of uniform extracted values in [0,1) (true value = 0.5)")
plt.axhline(y=0.5, color='r', linestyle='-')
plt.grid(True)
plt.show()
```

The expected values of the mean follows from the expression: $$ \left<x\right> = \int_0^1 x\rho(x)dx = \int_0^1 xdx = 0.5 $$

It can be noticed that the true value is in the range of 2$\sigma$ so the data are accettable.

```python
path = "1.1/output_stdv_1.1.out"

M, N = np.loadtxt(path, usecols=(0,1), delimiter=' ', unpack='true', max_rows=1)

ave, err = np.loadtxt(path, usecols=(0,1), delimiter=' ', unpack='true', skiprows=1)
  
L=int(M/N)

x = np.arange(N)
x*=L 

plt.errorbar(x,ave,yerr=err)
plt.xlabel('#throws')
plt.ylabel('<$\sigma$>')
plt.axhline(y=1/12, color='r', linestyle='-')
plt.title("Std. dev. of uniform extracted values in [0,1) (true value = 1/12)")
plt.grid(True)
plt.show()
```

The standard deviation instead follows form: $$ \sigma = \int_0^1 (x-\left<x\right>)^2\rho(x)dx = \frac{1}{12} $$

As in the graph of the mean value, the true value is in the edge of the errorbar so the data are accettable.

```python
path = "1.1/output_chisqrd_1.1.out"

chi_sqrd = np.loadtxt(path, usecols=(0), delimiter=' ', unpack='true')

x= np.arange(len(chi_sqrd))

k=100
x_chi = np.linspace(chi2.ppf(0.02,k), chi2.ppf(0.995,k), 100)

fig, (ax1,ax2) = plt.subplots(1, 2, figsize=(15,5))

ax1.hist(chi_sqrd, density=True, label="Obtained distribution")
ax1.plot(x_chi, chi2.pdf(x_chi, k), lw=2,  label=r'$\chi^2$ distribution per $k=100$')
ax1.set_xlabel('Bins')
ax1.set_ylabel('#$\chi^2$')
ax1.set_title("$\chi^2$ distribution")
ax1.legend()

ax2.plot(x,chi_sqrd)
ax2.set_xlabel('Repetitions')
ax2.set_ylabel('$\chi^2$')
ax2.set_title("$\chi^2$ values (expected value = 100)")

ax2.axhline(y=100, color='r', linestyle='-')

ax1.grid(True)
ax2.grid(True)

plt.show()
```

These graphs confirm the attendability of the tested random number generator.

### <span style="color:green"> Central Limit Theorem Test</span>

In this exercise we check the CLT, first using two disribution in accord to the ipotesis, then a Chauchy-Lorentz distribution with infinite variance.

```python
def gaussian(x, mean, amplitude, standard_deviation):
    return amplitude * np.exp( - (x - mean)**2 / (2*standard_deviation ** 2))

def CauchyLorentz(x,mean,gamma):
    return gamma/math.pi * 1/((x-mean)**2 + gamma**2)
```

```python
path = "1.2/output_UD_1.2.out"

N_bins = 75

N = np.loadtxt(path, usecols=(0), unpack='true', max_rows=1)

UD1, UD2, UD10, UD100 = np.loadtxt(path, usecols=(0,1,2,3), delimiter=' ', unpack='true', skiprows=1)

fig ,(ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(30, 6))

data_entries, bins = np.histogram(UD100, bins = N_bins, density=True)
binscenters = np.array([ (bins[i] + bins[i+1])/2 for i in np.arange(N_bins) ])

popt, pcov = curve_fit(gaussian, xdata=binscenters, ydata=data_entries)

ax4.plot(binscenters, gaussian(binscenters, *popt),color = 'firebrick', linewidth = 2)

ax1.hist(UD1, bins= N_bins, color ='papayawhip', density=True)
ax1.set_xlabel("Value")
ax1.set_ylabel("#counts")
ax1.set_title("Uniform Distribuition, N=1")

ax2.hist(UD2, bins= N_bins, color ='wheat', density=True)
ax2.set_xlabel("Value")
ax2.set_ylabel("#counts")
ax2.set_title("Uniform Distribuition, N=2")

ax3.hist(UD10, bins= N_bins, color = 'burlywood', density=True)
ax3.set_xlabel("Value")
ax3.set_ylabel("#counts")
ax3.set_title("Uniform Distribuition, N=10")

ax4.hist(UD100, bins= N_bins, color = 'goldenrod',density=True)
ax4.set_xlabel("Value")
ax4.set_ylabel("#counts")
ax4.set_title("Uniform Distribuition, N=100")

plt.show()
```

```python
path = "1.2/output_Exp_1.2.out"

N_bins=75

N = np.loadtxt(path, usecols=(0), unpack='true', max_rows=1)

Exp1, Exp2, Exp10, Exp100 = np.loadtxt(path, usecols=(0,1,2,3), delimiter=' ', unpack='true', skiprows=1)

fig ,(ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(30, 6))

data_entries, bins = np.histogram(Exp100, bins = N_bins, density=True)
binscenters = np.array([ (bins[i] + bins[i+1])/2 for i in np.arange(N_bins) ])

popt, pcov = curve_fit(gaussian, xdata=binscenters, ydata=data_entries)

ax4.plot(binscenters, gaussian(binscenters, *popt),color = 'dodgerblue', linewidth = 2)

ax1.hist(Exp1, bins= N_bins, color='azure', density=True)
ax1.set_xlabel("Value")
ax1.set_ylabel("#counts")
ax1.set_title("Exponential Distribuition, N=1")

ax2.hist(Exp2, bins= N_bins, color='lightcyan', density=True)
ax2.set_xlabel("Value")
ax2.set_ylabel("#counts")
ax2.set_title("Exponential Distribuition, N=2")

ax3.hist(Exp10, bins= N_bins, color='paleturquoise', density=True)
ax3.set_xlabel("Value")
ax3.set_ylabel("#counts")
ax3.set_title("Exponential Distribuition, N=10")

ax4.hist(Exp100, bins= N_bins, color ='mediumturquoise', density=True)
ax4.set_xlabel("Value")
ax4.set_ylabel("#counts")
ax4.set_title("Exponential Distribuition, N=100")

plt.show()
```

In the two previous cases, it can be noticed that the distribution converges to a Gaussian, due to the finite variance.

```python
path = "1.2/output_CL_1.2.out"

N_bins = 100
x_range = (-7.5,7.5)

N = np.loadtxt(path, usecols=(0), unpack='true', max_rows=1)

CL1, CL2, CL10, CL100 = np.loadtxt(path, usecols=(0,1,2,3), delimiter=' ', unpack='true', skiprows=1)

fig ,(ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(30, 6))

data_entries, bins = np.histogram(CL100, bins = N_bins, range = x_range, density=True)
binscenters = np.array([ (bins[i] + bins[i+1])/2 for i in np.arange(N_bins) ])

popt, pcov = curve_fit(CauchyLorentz, xdata=binscenters, ydata=data_entries)

ax4.plot(binscenters, CauchyLorentz(binscenters, *popt),color = 'indigo', linewidth = 2)

ax1.hist(CL1, bins= N_bins, range = x_range, color = "thistle", density=True)
ax1.set_xlabel("Value")
ax1.set_ylabel("#counts")
ax1.set_title("Cauchy-Lorentz Distribuition, N=1")

ax2.hist(CL2, bins= N_bins, range = x_range, color = 'plum', density=True)
ax2.set_xlabel("Value")
ax2.set_ylabel("#counts")
ax2.set_title("Cauchy-Lorentz Distribuition, N=2")

ax3.hist(CL10, bins= N_bins, range = x_range, color = 'orchid',density=True)
ax3.set_xlabel("Value")
ax3.set_ylabel("#counts")
ax3.set_title("Cauchy-Lorentz Distribuition, N=10")

ax4.hist(CL100, bins= N_bins, range = x_range, color = 'fuchsia', density=True)
ax4.set_xlabel("Value")
ax4.set_ylabel("#counts")
ax4.set_title("Cauchy-Lorentz Distribuition, N=100")

plt.show()
```

A Chauchy-Lorentz distribution is an attractor in the distribution space, so it remains unchanged with changing of N.

### <span style="color:green"> Buffon's Experiment </span>

The Buffon's experiment aim is to compute the value of $\pi$. To do this a lot of needle were launched on a grid, the fraction between the number of needle that touch a line and which not converges to $P=\frac{2L}{\pi d}$ where L is the lenght of the needle and d is the distance between the lines. So reversing that formula is possible to extimate $\pi$.

```python
path = "1.3/output_pi_1.3.out"

N, L = np.loadtxt(path, usecols=(0,1), delimiter=' ', unpack='true', max_rows=1)

pi, err = np.loadtxt(path, usecols=(0,1), delimiter=' ', unpack='true', skiprows=1)
  
x = np.arange(N)
x*=L 

plt.errorbar(x,pi,yerr=err)
plt.xlabel('#throws')
plt.ylabel('<$\pi$>')
plt.axhline(y=math.pi, color='r', linestyle='-')
plt.title("Buffon's experiment")
plt.grid(True)
plt.show()
```
